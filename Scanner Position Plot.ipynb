This notebook creates the plot showing scanner position over time.

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
def py_scanner_layer(scanner_range: int, time_step: int) -> int:
    """Calculates the position of a scanner within its range at a given time step."""
    cycle_midpoint = scanner_range - 1
    full_cycle = cycle_midpoint * 2
    cycle_position = time_step % full_cycle  
    return cycle_position if cycle_position <= cycle_midpoint else full_cycle - cycle_position
```

```python
SCANNER_DEPTH = 5
TIME_STEPS = np.arange((SCANNER_DEPTH - 1) * 2 + 1)
SCANNER_POS = np.int_([py_scanner_layer(SCANNER_DEPTH, t) for t in TIME_STEPS])
```

```python
with matplotlib.style.context('seaborn-talk'):
    fig, ax = plt.subplots(figsize=None)
    ax.plot(TIME_STEPS, SCANNER_POS, linestyle='dotted', color='gray')
    ax.scatter(TIME_STEPS, SCANNER_POS, s=500, marker='s', zorder=10)
    
    ax.set_xlabel('Time')
    ax.set_ylabel('Scanner Position')
    ax.set_xticks(range(9))
    ax.set_yticks(range(5))
    
    fig.savefig('./images/scanner_position.pdf')
```
